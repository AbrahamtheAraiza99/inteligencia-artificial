```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.DiGraph()

# Nodo principal
G.add_node("Seguridad en Redes")

# Nodos secundarios
G.add_edges_from([
    ("Seguridad en Redes", "Puertos No Utilizados"),
    ("Seguridad en Redes", "VLANs"),
    ("Seguridad en Redes", "Port Security"),
    ("Puertos No Utilizados", "Switch Cisco"),
    ("Puertos No Utilizados", "Comando shutdown"),
    ("Puertos No Utilizados", "Prevencin de accesos no autorizados"),
    ("Switch Cisco", "CLI"),
    ("CLI", "Comandos de configuracin"),
    ("Comando shutdown", "Desactivacin de FastEthernet0/1"),
    ("Desactivacin de FastEthernet0/1", "Evitar acceso de PC1"),
    ("VLANs", "Segmentacin Lgica"),
    ("Segmentacin Lgica", "Usuarios Administrativos"),
    ("Segmentacin Lgica", "Usuarios Comunes"),
    ("Port Security", "Bloqueo por MAC address"),
    ("Port Security", "Prevencin de dispositivos no autorizados")
])

# Dibujar la red semntica
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G, k=0.5, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="#B0E0E6", font_size=10, font_weight='bold', arrows=True, edge_color='gray')
plt.title("Red Semntica: Seguridad en Redes LAN", fontsize=14)
plt.show()
```
