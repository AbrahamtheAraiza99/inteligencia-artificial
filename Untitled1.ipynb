```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Base de datos de conocimientos con hechos y directivas
knowledge_base = pd.DataFrame({
    'Regla': ['Regla 1', 'Regla 2', 'Regla 3', 'Regla 4', 'Regla 5',
              'Regla 6', 'Regla 7', 'Regla 8', 'Regla 9', 'Regla 10'],
    'Condición': ['Fiebre', 'Dolor de cabeza', 'Tos seca', 'Dolor muscular', 'Fatiga',
                  'Nauseas', 'Escalofríos', 'Dolor en el pecho', 'Dificultad para respirar', 'Congestión nasal'],
    'Acción': ['Descansar y beber líquidos', 'Tomar analgésicos', 'Tomar jarabe para la tos', 'Realizar estiramientos',
               'Tomar agua y descansar', 'Tomar antieméticos', 'Vestirse en capas y descansar',
               'Buscar atención médica', 'Buscar atención médica inmediata', 'Usar descongestionante nasal'],
})

# Función para consultar el sistema experto
def consultar_sistema_experto(symptoms):
    recomendaciones = []
    for index, row in knowledge_base.iterrows():
        if row['Condición'] in symptoms:
            recomendaciones.append(row['Acción'])
    return recomendaciones

# Función para mostrar recomendaciones basadas en la selección
def mostrar_recomendaciones(*args):
    selected_symptoms = [symptom for symptom, widget in checkboxes.items() if widget.value]
    recomendaciones = consultar_sistema_experto(selected_symptoms)
    print("Recomendaciones:")
    for recomendacion in recomendaciones:
        print(f"- {recomendacion}")

# Crear los widgets de entrada (checkboxes para los síntomas)
checkboxes = {symptom: widgets.Checkbox(value=False, description=symptom) for symptom in knowledge_base['Condición']}

# Botón para ejecutar la consulta
button = widgets.Button(description="Obtener Recomendaciones")
button.on_click(mostrar_recomendaciones)

# Mostrar los checkboxes y el botón
display(widgets.VBox(list(checkboxes.values())))
display(button)
```
