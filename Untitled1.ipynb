```python

import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Base de datos de conocimientos con hechos y directivas
knowledge_base = pd.DataFrame({
    'Regla': ['Regla 1', 'Regla 2', 'Regla 3', 'Regla 4', 'Regla 5',
              'Regla 6', 'Regla 7', 'Regla 8', 'Regla 9', 'Regla 10'],
    'Condici√≥n': ['Fiebre', 'Dolor de cabeza', 'Tos seca', 'Dolor muscular', 'Fatiga',
                  'Nauseas', 'Escalofr√≠os', 'Dolor en el pecho', 'Dificultad para respirar', 'Congesti√≥n nasal'],
    'Acci√≥n': ['Descansar y beber l√≠quidos', 'Tomar analg√©sicos', 'Tomar jarabe para la tos', 'Realizar estiramientos',
               'Tomar agua y descansar', 'Tomar antiem√©ticos', 'Vestirse en capas y descansar',
               'Buscar atenci√≥n m√©dica', 'Buscar atenci√≥n m√©dica inmediata', 'Usar descongestionante nasal'],
})

# Funci√≥n para consultar el sistema experto
def consultar_sistema_experto(symptoms):
    recomendaciones = []
    for index, row in knowledge_base.iterrows():
        if row['Condici√≥n'] in symptoms:
            recomendaciones.append(row['Acci√≥n'])
    return recomendaciones

# Funci√≥n para mostrar recomendaciones basadas en la selecci√≥n
def mostrar_recomendaciones(*args):
    selected_symptoms = [symptom for symptom, widget in checkboxes.items() if widget.value]
    recomendaciones = consultar_sistema_experto(selected_symptoms)
    print("Recomendaciones:")
    for recomendacion in recomendaciones:
        print(f"- {recomendacion}")

# Crear los widgets de entrada (checkboxes para los s√≠ntomas)
checkboxes = {symptom: widgets.Checkbox(value=False, description=symptom) for symptom in knowledge_base['Condici√≥n']}

# Bot√≥n para ejecutar la consulta
button = widgets.Button(description="Obtener Recomendaciones")
button.on_click(mostrar_recomendaciones)

# Mostrar los checkboxes y el bot√≥n
display(widgets.VBox(list(checkboxes.values())))
display(button)
```
