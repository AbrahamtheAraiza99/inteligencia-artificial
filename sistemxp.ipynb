```python

import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Base de conocimientos: 10 reglas (síntomas y acciones)
knowledge_base = pd.DataFrame({
    'Regla': [f'Regla {i+1}' for i in range(10)],
    'Síntoma': ['Fiebre', 'Dolor de cabeza', 'Tos seca', 'Dolor muscular', 'Fatiga',
                'Náuseas', 'Escalofríos', 'Dolor en el pecho', 'Dificultad para respirar', 'Congestión nasal'],
    'Recomendación': ['Descansar y beber líquidos', 'Tomar analgésicos', 'Tomar jarabe para la tos', 'Realizar estiramientos',
                      'Tomar agua y descansar', 'Tomar antieméticos', 'Abrigarse y descansar',
                      'Buscar atención médica', 'Buscar atención médica urgente', 'Usar descongestionantes']
})

# Función del sistema experto
def consultar_sistema_experto(symptoms):
    recomendaciones = []
    for _, row in knowledge_base.iterrows():
        if row['Síntoma'] in symptoms:
            recomendaciones.append(f"- {row['Recomendación']}")
    return recomendaciones if recomendaciones else ["No se encontraron recomendaciones."]

# Función al presionar el botón
def mostrar_recomendaciones(b):
    clear_output(wait=True)  # Limpia la salida anterior
    selected = [s for s, cb in checkboxes.items() if cb.value]
    recomendaciones = consultar_sistema_experto(selected)

    print("✅ Recomendaciones del Sistema Experto:")
    for r in recomendaciones:
        print(r)

    # Volver a mostrar los widgets
    display(widgets.VBox(list(checkboxes.values())))
    display(button)

# Crear checkboxes para los síntomas
checkboxes = {symptom: widgets.Checkbox(value=False, description=symptom) for symptom in knowledge_base['Síntoma']}

# Botón para obtener recomendaciones
button = widgets.Button(description="Consultar Sistema Experto", button_style='success')
button.on_click(mostrar_recomendaciones)

# Mostrar en pantalla
display(widgets.VBox(list(checkboxes.values())))
display(button)
```
