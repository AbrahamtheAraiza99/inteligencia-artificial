```python
import nbformat
import os
import sys

def clean_widgets_metadata(notebook_path, output_path=None):
    if not os.path.isfile(notebook_path):
        print(f"‚ùå Archivo no encontrado: {notebook_path}")
        return

    # Leer el notebook
    nb = nbformat.read(notebook_path, as_version=nbformat.NO_CONVERT)
    modified = False

    # Limpiar metadata.widgets del notebook
    if 'widgets' in nb['metadata']:
        widgets_meta = nb['metadata']['widgets']
        if not isinstance(widgets_meta, dict) or 'state' not in widgets_meta:
            print(f"üßπ Limpiando 'metadata.widgets' del notebook: {notebook_path}")
            del nb['metadata']['widgets']
            modified = True

    # Limpiar metadata.widgets en celdas individuales (precauci√≥n)
    for cell in nb['cells']:
        if 'metadata' in cell and 'widgets' in cell['metadata']:
            widgets_meta = cell['metadata']['widgets']
            if not isinstance(widgets_meta, dict) or 'state' not in widgets_meta:
                print(f"üßπ Limpiando 'metadata.widgets' de una celda en: {notebook_path}")
                del cell['metadata']['widgets']
                modified = True

    # Guardar cambios
    if modified:
        output_file = output_path if output_path else notebook_path
        nbformat.write(nb, output_file)
        print(f"‚úÖ Notebook limpio guardado en: {output_file}")
    else:
        print(f"‚úîÔ∏è No se encontraron problemas en: {notebook_path}")

if __name__ == "__main__":
    if len(sys.argv) < 2:
        print("Uso: python clean_notebook_widgets.py archivo_notebook.ipynb [archivo_salida.ipynb]")
    else:
        input_file = sys.argv[1]
        output_file = sys.argv[2] if len(sys.argv) > 2 else None
        clean_widgets_metadata(input_file, output_file)

import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Base de conocimientos: 10 reglas (s√≠ntomas y acciones)
knowledge_base = pd.DataFrame({
    'Regla': [f'Regla {i+1}' for i in range(10)],
    'S√≠ntoma': ['Fiebre', 'Dolor de cabeza', 'Tos seca', 'Dolor muscular', 'Fatiga',
                'N√°useas', 'Escalofr√≠os', 'Dolor en el pecho', 'Dificultad para respirar', 'Congesti√≥n nasal'],
    'Recomendaci√≥n': ['Descansar y beber l√≠quidos', 'Tomar analg√©sicos', 'Tomar jarabe para la tos', 'Realizar estiramientos',
                      'Tomar agua y descansar', 'Tomar antiem√©ticos', 'Abrigarse y descansar',
                      'Buscar atenci√≥n m√©dica', 'Buscar atenci√≥n m√©dica urgente', 'Usar descongestionantes']
})

# Funci√≥n del sistema experto
def consultar_sistema_experto(symptoms):
    recomendaciones = []
    for _, row in knowledge_base.iterrows():
        if row['S√≠ntoma'] in symptoms:
            recomendaciones.append(f"- {row['Recomendaci√≥n']}")
    return recomendaciones if recomendaciones else ["No se encontraron recomendaciones."]

# Funci√≥n al presionar el bot√≥n
def mostrar_recomendaciones(b):
    clear_output(wait=True)  # Limpia la salida anterior
    selected = [s for s, cb in checkboxes.items() if cb.value]
    recomendaciones = consultar_sistema_experto(selected)

    print("‚úÖ Recomendaciones del Sistema Experto:")
    for r in recomendaciones:
        print(r)

    # Volver a mostrar los widgets
    display(widgets.VBox(list(checkboxes.values())))
    display(button)

# Crear checkboxes para los s√≠ntomas
checkboxes = {symptom: widgets.Checkbox(value=False, description=symptom) for symptom in knowledge_base['S√≠ntoma']}

# Bot√≥n para obtener recomendaciones
button = widgets.Button(description="Consultar Sistema Experto", button_style='success')
button.on_click(mostrar_recomendaciones)

# Mostrar en pantalla
display(widgets.VBox(list(checkboxes.values())))
display(button)
```
